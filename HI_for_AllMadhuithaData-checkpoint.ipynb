```python
"""
Implementing what i done for bearing to madhumitha's data
working prpperly - Done for only first column here
"""

import math
import glob
import os
import numpy as np
import pandas as pd
from IntelliMaint.spindle import Bearing  # Import Bearing class from bearing_diagnostics.py
#from IntelliMaint.intelliMaint_core import TimeDomain, AnomalyDetection, SOM, GPRDegradationModel
from IntelliMaint.data_analysis import SOM
from IntelliMaint.anomaly_detection import AnomalyDetection
from IntelliMaint.Training import  HealthIndicator

import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.fft import fft, fftfreq
from scipy.signal import cwt, find_peaks_cwt, ricker
from scipy.signal import butter, lfilter, savgol_filter, hilbert


# Set up file path and parameters
folder_path = r'C:\Users\DELL\Desktop\Madhumitha files\ALL_DATA/'
print(folder_path)
files = sorted(glob.glob(folder_path + '*.csv'))
print(len(files))
#files = sorted(glob.glob(r'C:\Users\DELL\Desktop\Template\data\2nd_test\2nd_test/'))
Bd = 0.331  # Ball diameter in inches
Pd = 2.815  # Pitch diameter in inches
Nb = 16     # Number of rolling elements
a = math.radians(15.17)  # Contact angle in radians
s = 2000 / 60  # Rotational frequency in Hz
sampRate = 25600  # Sampling rate in Hz    


# Initialize the Bearing instance
bearingOld = Bearing(files, Bd, Pd, Nb, a, s, sampRate)
print(f"Total files found: {len(files)}")

# Analyze raw data
bearingOld.analyze_raw_data(ber=0)  # Example: 0 indicates which column to analyze

# Extract features
n_cols = 3
feature_df = bearingOld.extract_features(ber=0, n_cols=n_cols, files=files)
print(feature_df.shape)

    
# Detect Anomaly
#health_indicator = HealthIndicator(feature_df)
#h_score, h_score_filtered, som, scaler_ = health_indicator.run_health_indicator()
#health_indicator.plot_results(h_score)


# Detect anomaly
normal_data_pts_som = 200
train_som = feature_df[:normal_data_pts_som]
test_som = feature_df
da = SOM() 
som, scaler_ = da.train(train_som)
mqe = da.predict(som, test_som, scaler_).reshape(-1, 1)

plt.figure(figsize=(16, 8))
plt.plot(mqe)

# Train Anomaly detection with normal health indicator
normal_data_pts_cosmo = 200
hi = mqe

hi_train = hi[normal_data_pts_som:normal_data_pts_som+normal_data_pts_cosmo]
hi_test = hi[normal_data_pts_som:]
ad = AnomalyDetection()
hi_train = pd.DataFrame(hi_train)
ad.train_cosmo(hi_train)


# Evaluate health score using health indicator test
hi_test = pd.DataFrame(hi_test)
health_score_test, _ = ad.test_cosmo(hi_test)
health_score_test = health_score_test.squeeze()

hi_train = pd.DataFrame(hi_train)
health_score_train, _ = ad.test_cosmo(hi_train)
health_score_train = health_score_train.squeeze()

score_train = []
score_test = []

for i in range(hi_train.shape[0]):
    score_train.append(health_score_train[i])

for i in range(hi_test.shape[0]):
    score_test.append(health_score_test[i])

h_score = score_train + score_test
h_score = np.array(h_score)

window_length_filter = 50
h_score_filtered = savgol_filter(h_score, window_length_filter, 1)
plt.figure(figsize=(20, 10))
plt.plot([i for i in range(100, h_score.shape[0])], h_score[100:h_score.shape[0]], label='hi')
plt.plot([i for i in range(100, h_score.shape[0])], h_score_filtered[100:h_score.shape[0]], label='hi_filtered')
plt.ylabel('health score')
plt.xlabel('samples')
plt.legend()
plt.show()


"""
    
deg_start_idx, score_till_incipient, initial_deg_pts = bearingOld.detect_anomaly(incipient_fault_threshold=0.6, normal_data_pts_som= normal_data_pts_som, files=files, ber=0, n_cols=n_cols, som=som, scaler_=scaler_, da=da, ad=ad)
    
    
 # Diagnostics: Prepare features and labels for diagnostics
feature_df = bearingOld.prepare_data1(feature_df, failure_idx=deg_start_idx, normal_data_pts_som=normal_data_pts_som, failure_mode="outer_race")
features, labels = bearingOld.prepare_data2(feature_df, failure_idx=deg_start_idx, normal_data_pts_som=normal_data_pts_som, failure_mode="outer_race")

 # Evaluate diagnostic features
diagnostic_features = bearingOld.evaluate_diagnostic_features(features, labels)
bearingOld.plot_diagnostic_features(diagnostic_features, top_n=5)
"""
```
